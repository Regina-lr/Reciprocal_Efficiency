```python
import nltools
import glob
import os
import pandas as pd
import nibabel as nb
import numpy as np
import seaborn as sns
%matplotlib inline
datadir ='/MVPA_for_Prediction'
```

### BrainMaps_3Conds

```python
file_list = glob.glob(os.path.join(datadir,'First_Level_3Conds','GIIN*.nii'))
file_list.sort()
file_list
```

```python
from nltools.mask import expand_mask, collapse_mask
from nltools.data import Brain_Data

Efficiency_data = Brain_Data(file_list)
```

```python
Efficiency_info = glob.glob(os.path.join(datadir,'First_Level_3Conds','Efficiency_Info.csv'))[0]
Efficiency_data.X = pd.read_csv(Efficiency_info)
```

```python
Efficiency_data.X.head()
```

```python
Communal_weight = Brain_Data(os.path.join(datadir,'Communal_concern_prediction_weight_map.nii.gz'))
Obligation_weight = Brain_Data(os.path.join(datadir,'Obligation_prediction_weight_map.nii.gz'))                            
```

### Key 

```python
predicted_Communal = Efficiency_data.similarity(Communal_weight,'dot_product')

dat_Communal = pd.DataFrame(data={'Subject':Efficiency_data.X['Subject'],'Predicted_PC_C':predicted_Communal,
                                  'Efficiency':Efficiency_data.X['Efficiency']})
```

```python
predicted_Obligation = Efficiency_data.similarity(Obligation_weight,'dot_product')

dat_Obligation = pd.DataFrame(data={'Subject':Efficiency_data.X['Subject'],'Predicted_PC_O':predicted_Obligation,
                                  'Efficiency':Efficiency_data.X['Efficiency']})
```

```python
dat_Obligation
```

```python
dat_Communal.to_csv(os.path.join(datadir,'Predicted_Communal_3Conds.csv'))
dat_Obligation.to_csv(os.path.join(datadir,'Predicted_Obligation_3Conds.csv'))
```

### Visualization

```python
with sns.plotting_context(context='paper',font_scale=2):
    sns.factorplot(data=dat_Communal,x='Efficiency',y='Predicted_PC_C')
```

```python
with sns.plotting_context(context='paper',font_scale=2):
    sns.factorplot(data=dat_Obligation,x='Efficiency',y='Predicted_PC_O')
```
